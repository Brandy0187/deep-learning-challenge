## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df.drop(columns=['EIN', 'NAME'], inplace=True)
```

```python
application_df.head()
```

```python
# Determine the number of unique values in each column.
unique_values_counts = application_df[
    ['APPLICATION_TYPE',
    'AFFILIATION',
    'CLASSIFICATION',
    'USE_CASE',
    'ORGANIZATION',
    'STATUS',
    'INCOME_AMT',
    'SPECIAL_CONSIDERATIONS',
    'ASK_AMT',
    'IS_SUCCESSFUL']
].apply(lambda x: x.nunique())

unique_values_counts.head()
```

```python
# Look at APPLICATION_TYPE value counts for binning
application_type_counts = application_df['APPLICATION_TYPE'].value_counts()
print(application_type_counts)
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
cutoff_value = 500
application_type_counts = application_df['APPLICATION_TYPE'].value_counts()
application_types_to_replace = application_type_counts[application_type_counts < cutoff_value].index.tolist()

print(application_types_to_replace)

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
classification_counts = application_df['CLASSIFICATION'].value_counts()
print(classification_counts)
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
classification_counts = application_df['CLASSIFICATION'].value_counts()
classification_counts_filtered = classification_counts[classification_counts > 1]
print(classification_counts_filtered)
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
cutoff_value = 200
classification_counts_filtered = application_df['CLASSIFICATION'].value_counts()
classifications_to_replace = classification_counts_filtered[classification_counts_filtered < cutoff_value].index.tolist()

print(classifications_to_replace)

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
appCategories = ['APPLICATION_TYPE', 'AFFILIATION', 'CLASSIFICATION', 'USE_CASE', 'ORGANIZATION', 'INCOME_AMT', 'STATUS', 'SPECIAL_CONSIDERATIONS', 'ASK_AMT', 'IS_SUCCESSFUL']
data_numeric = pd.get_dummies(application_df, columns=appCategories)
print(data_numeric)
```

```python
# Split our preprocessed data into our features and target arrays
features = data_numeric.drop(columns="APPLICATION_TYPE_Other")
target = data_numeric["APPLICATION_TYPE_Other"]

print(features)
print(target)

# Split the preprocessed data into a training and testing dataset
test_size = 0.2
random_seed = 42

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=test_size, random_state=random_seed)

print("Training data - Features shape:", X_train.shape)
print("Training data - Target shape:", y_train.shape)
print("Testing data - Features shape:", X_test.shape)
print("Testing data - Target shape:", y_test.shape)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit_transform(features)

# Scale the data
X_test_scaled = scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
#deep neural network - 10 inputs, 10 neurons

nnApplication = tf.keras.models.Sequential()

# First hidden layer
nnApplication.add(
    tf.keras.layers.Dense(input_dim=8793, units=10, activation='relu')
)

# Second hidden layer
nnApplication.add(
    tf.keras.layers.Dense(input_dim=8793, units=10, activation='relu')
)

# Output layer
nnApplication.add(
    tf.keras.layers.Dense(units=1, activation='sigmoid')
)

# Check the structure of the model
nnApplication.summary
```
